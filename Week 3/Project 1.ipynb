```python
import pandas as pd
import os

# CSV files to pull desired data
files = {
    "HA": "Hank Aaron Stats.csv",
    "BB": "Barry Bonds Stats.csv",
    "TC": "Ty Cobb Stats.csv",
    "WM": "Willie Mays Stats.csv",
    "SM": "Stan Musial Stats.csv",
    "AP": "Albert Pujols Stats.csv",
    "PR": "Pete Rose Stats.csv",
    "BR": "Babe Ruth Stats.csv",
    "HW": "Honus Wagner Stats.csv",
    "TW": "Ted Williams Stats.csv"
}

# Base directory path
base_dir = r"C:\Users\zscon\Desktop\SSE_591\repo\SSE591_Week3\Week 3\Baseball Statistics"

# Dictionary to store data frames for each player
players = {}

# Load data for each player into a pandas DataFrame
for player, file_name in files.items():
    file_path = os.path.join(base_dir, file_name)
    players[player] = pd.read_csv(file_path)

# Dictionary to store total home runs for each year
total_home_runs = {}

# Collect information on number of Homeruns for all players
for player, data in players.items():
    print(f"\n{player} Home Runs of all players:")
    player_data = data[data['HR']]
    print(player_data['HR'])

    # Sort player_data to show Homerun total for career
    player_data_sort = player_data.sort_values(by='HR', ascending=False)

# Collect information on number of Hits for all players
for player, data in players.items():
    print(f"\n{player} Hits of all players:")
    player_data = data[data['Hits']]
    print(player_data['Hits'])

    # Sort player_data to show Hit total for career
    player_data_sort = player_data.sort_values(by='Hits', ascending=False)

# Collect information on number of RBI for all players
for player, data in players.items():
    print(f"\n{player} RBIs of all players:")
    player_data = data[data['RBI']]
    print(player_data['RBI'])

    # Sort player_data to show RBI total for career
    player_data_sort = player_data.sort_values(by='RBI', ascending=False)
```
